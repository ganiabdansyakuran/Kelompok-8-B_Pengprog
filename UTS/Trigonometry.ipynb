# Trigonometri
**Trigonometri** adalah cabang dari matematika yang mempelajari hubungan antara sudut dan panjang sisi dalam segitiga. Kata *trigonometri* berasal dari bahasa Yunani, yaitu *trigonon* yang berarti "tiga sudut", dan *metron* yang berarti "mengukur". 

Ilmu ini sangat penting dalam berbagai bidang seperti astronomi, navigasi, geodesi, fisika, serta cabang ilmu kebumian seperti geofisika. Misalnya, dalam geofisika, trigonometri membantu menentukan jarak antara dua titik di permukaan bumi dan arah getaran gelombang seismik.

Sementara itu, **Trigonometri Bola** merupakan pengembangan dari trigonometri yang diterapkan pada permukaan bola. Ini sangat penting untuk menghitung sudut dan jarak di permukaan bumi, terutama dalam sistem koordinat lintang dan bujur.

Trigonometri memiliki peran penting dalam berbagai bidang ilmu dan kehidupan sehari-hari, seperti dalam bidang teknik, fisika, astronomi, arsitektur, bahkan dalam teknologi modern seperti grafik komputer dan sistem navigasi GPS. Melalui trigonometri, kita dapat menghitung jarak, tinggi, dan sudut yang tidak dapat diukur secara langsung hanya dengan menggunakan konsep dasar dan identitas trigonometri.

## Fungsi Trigonometri Dasar

Fungsi trigonometri menghubungkan sudut dengan rasio sisi-sisi dalam segitiga siku-siku. Ada enam fungsi utama:

1. $\sin$
2. $\cos$
3. $\tan$
4. $\csc$
5. $\sec$
6. $\cot$

# Segitiga Siku-siku
![segitiga](segitiga.png)

Segitiga siku-siku adalah dasar penting dalam materi trigonometri, karena semua fungsi trigonometri (seperti sinus, cosinus, dan tangen) awalnya didefinisikan berdasarkan segitiga jenis ini.
Segitiga siku-siku adalah segitiga yang memiliki satu sudut sebesar 90° (siku-siku). Sisi-sisi segitiga ini terdiri dari:
1. **Sisi miring (hipotenusa)**: sisi terpanjang yang berhadapan langsung dengan sudut siku-siku.
2. **Sisi depan (opposite)**: sisi yang berada di depan sudut yang sedang dibahas.
3. **Sisi samping (adjacent)**: sisi yang berada di samping sudut (selain sisi miring).

# Fungsi Trigonometri Dasar

Untuk sebuah segitiga siku-siku dengan sudut $( \theta )$, terdapat tiga fungsi trigonometri dasar yang digunakan untuk menghitung hubungan antara sisi-sisi segitiga:

1. **Sinus (sin)**
   $$
   \sin(\theta) = \frac{\text{sisi depan}}{\text{sisi miring}} \tag{1}
   $$

2. **Kosinus (cos)**
   $$
   \cos(\theta) = \frac{\text{sisi samping}}{\text{sisi miring}}\tag{2}
   $$

3. **Tangen (tan)**
   $$
   \tan(\theta) = \frac{\text{sisi depan}}{\text{sisi samping}}\tag{3}
   $$

4. **Sekan (sec)**
$$
\sec(\theta) = \frac{1}{cos(\theta)}\tag{4}
$$

5. **Kosekan (csc)**
$$
\csc(\theta) = \frac{1}{cos(\theta)}\tag{5}
$$

6. **Kotangen (cot)**
$$
\cot(\theta) = \frac{1}{tan(\theta)} = \frac{cos(\theta)}{sin(\theta)}\tag{6}
$$

# Identitas Trigonometri

Identitas trigonometri adalah persamaan-persamaan penting yang berlaku untuk semua nilai sudut.

## Identitas Pythagoras

$$
\sin^2 \theta + \cos^2 \theta = 1 \tag{7}
$$

Turunan dari identitas Pythagoras:

$$
1 + \tan^2 \theta = \sec^2 \theta \tag{8}
$$

$$
1 + \cot^2 \theta = \csc^2 \theta \tag{9}
$$

## Identitas Penjumlahan dan Sudut Ganda

- Penjumlahan sudut:
  $$
  \sin(a \pm b) = \sin a \cos b \pm \cos a \sin b \tag{10}
  $$

  $$
  \cos(a \pm b) = \cos a \cos b \mp \sin a \sin b \tag{11}
  $$

- Sudut ganda:
  $$
  \sin(2a) = 2 \sin a \cos a \tag{12}
  $$

  $$
  \cos(2a) = \cos^2 a - \sin^2 a \tag{13}
  $$

Identitas ini sering digunakan dalam menyelesaikan persamaan trigonometri yang kompleks, atau dalam bidang fisika dan teknik seperti analisis gelombang dan vibrasi.

# Trigonometri Bola

Trigonometri bola mempelajari hubungan sudut dan sisi pada permukaan bola, banyak digunakan dalam ilmu bumi dan astronomi. Berbeda dengan trigonometri bidang, trigonometri bola lebih kompleks karena bekerja pada permukaan lengkung.

## Lingkaran Besar
Jarak terpendek antar dua titik pada bola adalah **busur lingkaran besar**, yaitu lingkaran yang membagi bola menjadi dua bagian sama luas. Contoh: khatulistiwa dan garis bujur adalah lingkaran besar. Dua lingkaran besar selalu bertemu di dua titik.

## Segitiga Bola
Segitiga bola adalah segitiga yang dibentuk oleh tiga busur lingkaran besar pada permukaan bola. Sisi-sisinya (disebut $a$, $b$, $c$) adalah busur lingkaran besar, dan sudut-sudutnya ($A$, $B$, $C$) adalah sudut dihedral, yaitu sudut antara dua bidang lingkaran besar yang bertemu di titik sudut.

Berbeda dengan segitiga bidang:

Jumlah sudut segitiga bidang selalu 180° ($\pi$ radian).

Jumlah sudut segitiga bola selalu lebih dari 180°, bisa 200° atau lebih, tergantung ukuran segitiga.

Luas segitiga bola dihitung dengan rumus:

$$ S = \frac{\pi R^2}{180^\circ} (A + B + C - 180^\circ) \tag{14} $$

Di mana:

$S$: luas segitiga bola.

$R$: jari-jari bola (misalnya, jari-jari Bumi).

$A$, $B$, $C$: sudut dihedral dalam derajat.

Rumus ini menunjukkan bahwa luas bergantung pada seberapa besar jumlah sudut melebihi 180°.

## Sudut Trihedral

Selain sudut permukaan ($A, B, C$), segitiga bola juga memiliki sudut trihedral ($\alpha, \beta, \gamma$) yang dibentuk oleh tiga garis radial dari pusat bola ke titik-titik sudut segitiga . Panjang sisi segitiga bola ($a, b, c$) sebenarnya adalah panjang busur lingkaran besar yang menghubungkan titik-titik sudut tersebut, dan panjang ini diperoleh dengan mengalikan sudut trihedral (dalam radian) dengan jari-jari bola ($R$). Dalam rumus-rumus trigonometri bola, sisi-sisi segitiga seringkali dinyatakan dalam bentuk sudut trihedral yang bersesuaian.

**Aturan Sinus untuk Segitiga Bola:**

$$\frac{\sin a}{\sin A} = \frac{\sin b}{\sin B} = \frac{\sin c}{\sin C} \quad \quad \tag{15}$$
**Hukum Kosinus Pertama (untuk Sudut Trihedral):**

$$\cos \alpha = \cos \beta \cos \gamma + \sin \beta \sin \gamma \cos A \quad \quad \tag{16}$$

## Aturan Sinus untuk Segitiga Bola:

$$\frac{\sin a}{\sin A} = \frac{\sin b}{\sin B} = \frac{\sin c}{\sin C} \quad \quad \tag{17}$$
## Hukum Kosinus Pertama (untuk Sudut Trihedral):

$$\cos \alpha = \cos \beta \cos \gamma + \sin \beta \sin \gamma \cos A \quad \quad \tag{18}$$

## Hukum Kosinus Kedua (untuk Sudut Permukaan):

$$\cos A = -\cos B \cos C + \sin B \sin C \cos \alpha \quad \quad \tag{19}$$

Untuk segitiga bola siku-siku (dengan $C = 90^\circ$):

$$\cos A = \sin B \cos a \quad \quad \tag{20}$$

## Kurva Waktu Tempuh Gelombang Seismik

Gelombang seismik berperilaku serupa dengan cahaya, yaitu mengikuti hukum **Snell** ketika dibiaskan di antarmuka dua lapisan yang memiliki kecepatan dan densitas yang berbeda. Ini berlaku untuk kedua jenis gelombang, yaitu **gelombang P** (primer) dan **gelombang S** (sekunder). Namun, pada gelombang seismik ada hal yang sedikit lebih rumit, karena adanya **konversi mode**, yang berarti gelombang P yang datang akan menghasilkan kedua jenis gelombang (P dan S) pada antarmuka, begitu pula sebaliknya.

Sebagai contoh, saat sebuah gelombang P datang dengan sudut tertentu, gelombang tersebut akan dipantulkan kembali dengan sudut yang sama. Namun, untuk gelombang S yang terbentuk, sudut pantulannya bergantung pada hubungan antara kecepatan gelombang P dan gelombang S.

Jika kita menuliskan persamaan ini secara matematis:

$$
\frac{\sin \theta_i}{v_{P1}} = \frac{\sin \theta_2}{v_{S1}} \tag{7}
$$

Di mana:
- $( \theta_i )$ adalah sudut datang gelombang P,
- $( \theta_2 )$ adalah sudut pantul untuk gelombang S,
- $( v_{P1} )$ dan $( v_{S1} )$ adalah kecepatan gelombang P dan S pada lapisan pertama.

Karena **gelombang P** bergerak lebih cepat daripada **gelombang S** ($( v_{P1} > v_{S1} )$), maka sudut pantul gelombang S lebih kecil dibandingkan dengan sudut pantul gelombang P.

Hukum Snell untuk refraksi gelombang P dan S dapat ditulis sebagai berikut:

$$
\frac{\sin \theta_i}{\sin \psi_P} = \frac{v_{P1}}{v_{P2}}, \quad
\frac{\sin \theta_i}{\sin \psi_S} = \frac{v_{P1}}{v_{S2}} \tag{8}
$$

Dimana:
- $( \psi_P )$ adalah sudut refraksi untuk gelombang P,
- $( \psi_S )$ adalah sudut refraksi untuk gelombang S.

Sebagai hasilnya, gelombang P dan S akan memiliki sudut refraksi yang berbeda karena kecepatan gelombang yang berbeda. Jika sudut datang ( $( \theta )$ ) meningkat, maka sudut refraksi ( $( \psi )$ ) juga akan meningkat. Ada yang disebut sebagai **sudut kritis**, yaitu sudut yang terjadi ketika gelombang refraksi mencapai $( 90^\circ )$, di mana gelombang tersebut akan bergerak sepanjang antarmuka antara dua lapisan. Untuk gelombang P, sudut kritis ini bisa dihitung dengan persamaan berikut:

$$
\sin \theta_c = \frac{v_{P1}}{v_{P2}} \Rightarrow \theta_c = \sin^{-1} \left( \frac{v_{P1}}{v_{P2}} \right) \tag{9, 10}
$$

Namun, ini hanya berlaku jika $( v_{P1} < v_{P2} )$. Persamaan ini juga berlaku untuk gelombang S dengan cara yang serupa.

## Contoh Soal

### Soal 1
Diketahui sebuah segitiga ABC dengan sudut ∠A = 30°, ∠B = 60°, dan sisi BC = 10 cm. Hitung panjang sisi AB dan AC menggunakan hukum sinus.

```python
import math

angle_A_deg = 30  # Sudut A dalam derajat
angle_B_deg = 60  # Sudut B dalam derajat
side_c = 10       # Panjang sisi BC

angle_C_deg = 180 - angle_A_deg - angle_B_deg

angle_A_rad = math.radians(angle_A_deg)
angle_B_rad = math.radians(angle_B_deg)
angle_C_rad = math.radians(angle_C_deg)

side_a = (side_c * math.sin(angle_A_rad)) / math.sin(angle_C_rad)
side_b = (side_c * math.sin(angle_B_rad)) / math.sin(angle_C_rad)

print(f"Panjang sisi AB: {side_a:.2f} cm")
print(f"Panjang sisi AC: {side_b:.2f} cm")
```

## Soal 2

Diketahui sebuah segitiga siku-siku ABC dengan:

Sisi alas (adjacent) AC = 3 cm

Sisi depan (opposite) BC = 4 cm

Sudut siku-siku di C

Tentukan:

1. Nilai $\sin A$, $\cos A$, dan $\tan A$

2. Panjang hipotenusa AB

```python
import math

adjacent = 3  # AC
opposite = 4  # BC

hypotenuse = math.hypot(adjacent, opposite)

sin_A = opposite / hypotenuse
cos_A = adjacent / hypotenuse
tan_A = opposite / adjacent

print(f"AB (hipotenusa) = {hypotenuse:.2f} cm")
print(f"sin(A) = {sin_A:.3f}")
print(f"cos(A) = {cos_A:.3f}")
print(f"tan(A) = {tan_A:.3f}")
```
