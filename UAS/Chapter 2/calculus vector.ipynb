# 2.1 Derivatives (Turunan)

## Pengertian Turunan

Dalam kalkulus, turunan menggambarkan beberapa cepat suatu nilai berubah terhadap variabel lainnya. Secara intuitif, jika kita memiliki sebuah grafik, maka turunan adalah kemiringan garis singgung di titik tertentu.

Untuk sebuah fungsi skalar \( f(x) \), turunan terhadap \( x \) dituliskan sebagai:

$$ f'(x) = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x} $$

Ini berarti kita melihat bagaimana nilai \( f(x) \) berubah saat \( x \) berubah sangat kecil.

## Turunan dari Fungsi Vektor

Jika kita punya fungsi vektor posisi $$ \vec{r}(t) = (x(t), y(t), z(t)) $$, maka turunan dihitung pada tiap komponennya:

$$ \frac{d\vec{r}}{dt} = \left( \frac{dx}{dt}, \frac{dy}{dt}, \frac{dz}{dt} \right) $$

Turunan ini mengasilkan vektor baru yang menggambarkan arah perubahan posisi seiring waktuâ€”yaitu vektor kecepatan.

## Contoh Soal

Misalkan \( f(t) = t^3 - 2t^2 + 1 \). Hitung turunan pertama \( f'(t) \) dan gambarkan grafik fungsi serta turunannya.

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-2, 4, 100)
f = t**3 - 2*t**2 + 1
df = 3*t**2 - 4*t  # Turunan dari f(t)

plt.plot(t, f, label='f(t)', color='blue')
plt.plot(t, df, label="f'(t)", linestyle='--', color='red')
plt.xlabel('t')
plt.ylabel('Nilai')
plt.title('Grafik f(t) dan f\'(t)')
plt.axhline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```

## 2.2 Vector Functions (Fungsi Vektor)

### Apa itu Fungsi Vektor?

Fungsi vektor adalah fungsi yang keluaranannya berupa vektor, bukan angka biasa. Contohnya:

$$ \vec{r}(t) = (x(t), y(t), z(t)) $$

Artinya, untuk setiap nilai \( t \), kita mendapatkan sebuah titik dalam ruang 3 dimensi. Fungsi vektor banyak digunakan untuk menggambarkan lintasan (trajectory) suatu objek yang bergerak.

### Turunan dari Fungsi Vektor

Seperti sebelumnya, kita bisa menurunkan fungsi vektor dengan menurunkan tiap komponennya. Hasilnya adalah vektor yang menunjukkan arah dan kecepatan perubahan posisi, yang akan dibahas lebih dalam di subbab selanjutnya.

### Contoh Soal

Misalkan sebuah partikel memiliki posisi:

$$ \vec{r}(t) = (t, t^2, t^3) $$

Tampilkan lintasan objek tersebut dalam ruang 3D.

```python
from mpl_toolkits.mplot3d import Axes3D

t = np.linspace(0, 2, 100)
x = t
y = t**2
z = t**3

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, label='Lintasan r(t)', color='green')
ax.set_xlabel('x = t')
ax.set_ylabel('y = t^2')
ax.set_zlabel('z = t^3')
ax.set_title('Lintasan Partikel di Ruang 3D')
ax.legend()
plt.show()
```

## 2.3 Velocity and Acceleration (Kecepatan dan Percepatan)

### Kecepatan

Kecepatan adalah turunan pertama dari posisi terhadap waktu. Jika posisi partikel dituliskan sebagai \(\vec{r}(t)\), maka:

$$ \vec{v}(t) = \frac{d\vec{r}}{dt} $$

Artinya, vektor kecepatan menunjukkan kemana dan seberapa cepat partikel tersebut bergerak.

### Percepatan

Percepatan adalah turunan dari kecepatan, atau turunan kedua dari posisi:

$$ \vec{a}(t) = \frac{d\vec{v}}{dt} = \frac{d^2\vec{r}}{dt^2} $$

Percepatan menggambarkan bagaimana kecepatan berubah terhadap waktu.

### Contoh Soal

Diberikan $$ \vec{r}(t) = (t, t^2, \sin(t)) $$, hitung kecepatan dan percepatannya.

```python
t = np.linspace(0, 2*np.pi, 200)
r = np.array([t, t**2, np.sin(t)])

# Turunan pertama (kecepatan)
v = np.array([np.ones_like(t), 2*t, np.cos(t)])

# Turunan kedua (percepatan)
a = np.array([np.zeros_like(t), 2*np.ones_like(t), -np.sin(t)])

# Grafik kecepatan
plt.plot(t, v[0], label='v_x(t)', color='blue')
plt.plot(t, v[1], label='v_y(t)', color='orange')
plt.plot(t, v[2], label='v_z(t)', color='green')
plt.title('Komponen Kecepatan')
plt.xlabel('t')
plt.ylabel('Nilai')
plt.grid(True)
plt.legend()
plt.show()

# Grafik percepatan
plt.plot(t, a[0], label='a_x(t)', linestyle='--')
plt.plot(t, a[1], label='a_y(t)', linestyle='--')
plt.plot(t, a[2], label='a_z(t)', linestyle='--')
plt.title('Komponen Percepatan')
plt.xlabel('t')
plt.ylabel('Nilai')
plt.grid(True)
plt.legend()
plt.show()
```
