# 2.1 TURUNAN

## Definisi Turunan

Turunan fungsi \( y = f(x) \) di titik \( x \) mengukur laju perubahan fungsi tersebut. Secara matematis didefinisikan sebagai:

$$
\frac{d}{dx} f(x) = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x} \tag{2.1.1}
$$

### Penjelasan:
- $$Limit\ \lim_{\Delta x \to 0}\ menunjukkan\ pendekatan\ perubahan\ yang\ sangat\ kecil$$
- $$Pembilang\ (f(x + \Delta x) - f(x))\ merepresentasikan\ perubahan\ nilai\ fungsi\$$
- $$Penyebut\ \Delta x\ adalah\ perubahan\ variabel\ independen$$
- $$Hasilnya\ adalah\ kemiringan\ garis\ singgung\ di\ titik\ (x)$$

### Pendekatan Numerik
Dalam komputasi, limit sulit diimplementasikan sehingga digunakan pendekatan:

#### 1. Selisih Maju (Forward Difference)
$$
\frac{f(x + h) - f(x)}{h} \tag{2.1.2a}
$$
- Menggunakan nilai fungsi di \(x\) dan titik di depan (\(x + h\))
- Error proporsional terhadap \(h\)

#### 2. Selisih Mundur (Backward Difference)
$$
\frac{f(x) - f(x - h)}{h} \tag{2.1.2b}
$$
- Menggunakan nilai fungsi di \(x\) dan titik di belakang (\(x - h\))
- Error proporsional terhadap \(h\)

#### 3. Selisih Pusat (Central Difference)
$$
\frac{f(x + h) - f(x - h)}{2h} \tag{2.1.2c}
$$
- Menggabungkan titik di depan dan belakang
- Error proporsional terhadap \(h^2\) (lebih akurat)
- Merupakan rata-rata dari selisih maju dan mundur

### Turunan Parsial
Untuk fungsi multi-variabel \( f(x, y) \), turunan parsial mengukur laju perubahan terhadap satu variabel sambil mempertahankan variabel lain tetap konstan.

#### Terhadap variabel \(x\):
$$
\frac{\partial}{\partial x} f(x, y) = \lim_{\Delta x \to 0} \frac{f(x + \Delta x, y) - f(x, y)}{\Delta x} \tag{2.1.3a}
$$
- Variabel \(y\) diperlakukan konstan
- Hanya \(x\) yang mengalami perubahan

#### Terhadap variabel \(y\):
$$
\frac{\partial}{\partial y} f(x, y) = \lim_{\Delta y \to 0} \frac{f(x, y + \Delta y) - f(x, y)}{\Delta y} \tag{2.1.3b}
$$
- Variabel \(x\) diperlakukan konstan
- Hanya \(y\) yang mengalami perubahan

#### Karakteristik:
1. $$Simbol\ \partial\ menunjukkan\ turunan\ parsial$$
2. $$Dalam\ perhitungan\ praktis,\ nilai\ (h)\ dipilih\ cukup\ kecil\ (biasanya\ (10^{-5})\ hingga\ (10^{-8}))$$
3. $$Pendekatan\ selisih\ yang\ sama\ dapat\ diaplikasikan\ untuk\ turunan\ parsial$$

## Aturan Rantai (Chain Rule)

### Untuk Fungsi Satu Variabel
Jika \(y = f(u)\) terdiferensialkan terhadap \(u\) dan \(u = g(x)\) merupakan fungsi dari variabel \(x\), maka turunan \(y\) terhadap \(x\) adalah:
$$
\frac{dy}{dx} = \frac{dy}{du} \cdot \frac{du}{dx} \tag{2.1.4}
$$

### Untuk Fungsi Dua Variabel
Jika \(z = f(u,v)\) terdiferensialkan terhadap \(u\) dan \(v\), dengan \(u = g(x,y)\) dan \(v = h(x,y)\) memiliki turunan parsial kontinu, maka:
$$
\frac{\partial z}{\partial x} = \frac{\partial z}{\partial u} \frac{\partial u}{\partial x} + \frac{\partial z}{\partial v} \frac{\partial v}{\partial x} \tag{2.1.5a}
$$
$$
\frac{\partial z}{\partial y} = \frac{\partial z}{\partial u} \frac{\partial u}{\partial y} + \frac{\partial z}{\partial v} \frac{\partial v}{\partial y} \tag{2.1.5b}
$$

**Inti Aturan Rantai:**  
Turunan fungsi komposit = perkalian turunan tiap "rantai" variabel perantara


## Gradien (Gradient)

### Definisi Berdasarkan Dimensi
1. **Fungsi Satu Variabel** \((f(x))\):
$$
\nabla f(x) = \mathbf{u}_x \frac{\partial f}{\partial x} \tag{2.1.6}
$$

2. **Fungsi Dua Variabel** \((f(x,y))\):
$$
\nabla f(x,y) = \mathbf{u}_x \frac{\partial f}{\partial x} + \mathbf{u}_y \frac{\partial f}{\partial y} \tag{2.1.7}
$$

3. **Fungsi Tiga Variabel** \((f(x,y,z))\):
$$
\nabla f(x,y,z) = \mathbf{u}_x \frac{\partial f}{\partial x} + \mathbf{u}_y \frac{\partial f}{\partial y} + \mathbf{u}_z \frac{\partial f}{\partial z} \tag{2.1.8}
$$

### Operator Nabla (âˆ‡)
Operator vektor diferensial dalam koordinat Kartesius 3D:
$$
\nabla = \mathbf{u}_x \frac{\partial}{\partial x} + \mathbf{u}_y \frac{\partial}{\partial y} + \mathbf{u}_z \frac{\partial}{\partial z} \tag{2.1.9}
$$

**Sifat Penting Gradien:**  
- Menunjukkan arah kenaikan fungsi paling curam  
- Besarnya = laju perubahan maksimum  
- Selalu tegak lurus terhadap permukaan level

## Turunan Berarah (Directional Derivative)

### Definisi
Turunan fungsi f(x,y,z) searah vektor satuan $$\mathbf{u} = \alpha_x \mathbf{u}_x + \alpha_y \mathbf{u}_y + \alpha_z \mathbf{u}_z\ (\|\mathbf{u}\|=1):$$
$$
\nabla_{\mathbf{u}} f(x,y,z) = \lim_{t \to 0} \frac{f(x+t\alpha_x, y+t\alpha_y, z+t\alpha_z) - f(x,y,z)}{t} \tag{2.1.10a}
$$

### Hubungan dengan Gradien
Dapat dihitung sebagai produk titik gradien dan vektor arah:
$$
\nabla_{\mathbf{u}} f(x,y,z) = \nabla f(x,y,z) \cdot \mathbf{u} \tag{2.1.10b}
$$

**Karakteristik:**  
- $$Maksimum\ ketika\ \mathbf{u}\ searah\ dengan\ \nabla f\$$  
- $$Minimum\ ketika\ \mathbf{u}\ berlawanan\ arah\ dengan\ \nabla f\$$  
- $$Nol\ ketika\ \mathbf{u}\ tegak\ lurus\ dengan\ \nabla f\$$

## Garis Singgung dan Bidang Singgung

### Untuk Kurva \(y = f(x)\)
Persamaan garis singgung di titik \((x_1, y_1)\):
$$
y = y_1 + \left. \frac{\partial f}{\partial x} \right|_{x=x_1} (x - x_1) \tag{2.1.11}
$$

### Untuk Kurva Level \(f(x,y) = c\)
Persamaan garis singgung di titik \((x_1, y_1)\):
$$
\left. \frac{\partial f}{\partial x} \right|_{(x_1,y_1)} (x - x_1) + \left. \frac{\partial f}{\partial y} \right|_{(x_1,y_1)} (y - y_1) = 0 \tag{2.1.12}
$$

### Untuk Permukaan Level \(f(x,y,z) = c\)
Persamaan bidang singgung di titik \((x_1, y_1, z_1)\):
$$
\left. \frac{\partial f}{\partial x} \right|_{(x_1,y_1,z_1)} (x - x_1) + \left. \frac{\partial f}{\partial y} \right|_{(x_1,y_1,z_1)} (y - y_1) + \left. \frac{\partial f}{\partial z} \right|_{(x_1,y_1,z_1)} (z - z_1) = 0 \tag{2.1.13}
$$

**Interpretasi Geometris:**  
- Gradien \(\nabla f\) berperan sebagai vektor normal  
- Persamaan (2.1.12) dan (2.1.13) merupakan perluasan konsep garis singgung ke dimensi tinggi  
- Persamaan (2.1.11) adalah kasus khusus dari (2.1.12) dengan \(f(x,y) = y - f(x) = 0\)

## Example

### Konsep Dasar Gradien

**Gradien** (âˆ‡f) adalah vektor yang:
- Menunjukkan arah **peningkatan tercepat** nilai fungsi
- Komponennya terdiri dari turunan parsial fungsi terhadap setiap variabel
- Untuk fungsi 2D: $\nabla f = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right)$
- Untuk fungsi 3D: $\nabla f = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y}, \frac{\partial f}{\partial z} \right)$

**Sifat penting gradien**:
1. Selalu tegak lurus terhadap kurva/permukaan tingkat
2. Besarnya menunjukkan laju perubahan fungsi
3. Arahnya menunjukkan orientasi kemiringan maksimum

---

### Contoh 2.1(a): Kurva Tingkat 2D

#### Persamaan Kurva Tingkat
$$ f(x, y) = x^2 - y = 4 \quad \text{(E2.1a.1)} $$

Kurva tingkat adalah himpunan titik-titik (x,y) di mana fungsi bernilai konstan (f(x,y) = 4). Contoh titik pada kurva ini adalah pâ‚ = (2,0).

#### Menghitung Gradien di Titik pâ‚ = (2,0)

1. Hitung turunan parsial:
   - $\frac{\partial f}{\partial x} = 2x$
   - $\frac{\partial f}{\partial y} = -1$

2. Evaluasi di titik (2,0):
   - $\frac{\partial f}{\partial x}\bigg|_{(2,0)} = 2(2) = 4$
   - $\frac{\partial f}{\partial y}\bigg|_{(2,0)} = -1$

3. Bentuk vektor gradien:
$$ \nabla f(2,0) = 4\mathbf{u}_x - \mathbf{u}_y \quad \text{(E2.1a.2)} $$

**Interpretasi**:
- Komponen x positif (4) artika fungsi meningkat jika bergerak ke kanan
- Komponen y negatif (-1) artinya fungsi menurun jika bergerak ke atas
- Arah vektor: kanan-bawah (sudut â‰ˆ -14Â° terhadap sumbu x)

#### Menentukan Persamaan Garis Singgung

Garis singgung di titik pâ‚ harus:
1. Melewati titik (2,0)
2. Tegak lurus terhadap vektor gradien (4, -1)

Persamaan umum garis singgung:
$$ \nabla f \cdot (\mathbf{r} - \mathbf{r}_0) = 0 $$
$$ \frac{\partial f}{\partial x}(x - x_1) + \frac{\partial f}{\partial y}(y - y_1) = 0 $$

Substitusi nilai:
$$ 4(x - 2) + (-1)(y - 0) = 0 $$
$$ 4x - 8 - y = 0 $$
$$ y = 4x - 8 \quad \text{(E2.1a.3)} $$

**Verifikasi**:
- Gradien garis: m = 4
- Gradien vektor normal: (4,-1) â†’ kemiringan tegak lurus = -1/(4) = -0.25
- Hasil kali kemiringan: 4 Ã— (-0.25) = -1 â†’ terbukti tegak lurus

#### Visualisasi (Gambar 2.1a dan 2.1b)

- **Gambar 2.1a**: 
  - Kurva parabola $y = x^2 - 4$
  - Garis singgung $y = 4x - 8$ di titik (2,0)
  - Vektor gradien (4,-1) yang tegak lurus garis singgung

- **Gambar 2.1b**:
  - Kumpulan kurva tingkat untuk c = -1,0,...,5
  - Vektor-vektor gradien di berbagai titik grid
  - Teramati bahwa gradien selalu tegak lurus kontur

---

### Contoh 2.1(b): Permukaan Tingkat 3D

#### Persamaan Permukaan Tingkat
$$ f(x, y, z) = x^2 - y - 4z = 0 \quad \text{(E2.1b.1)} $$

Permukaan tingkat adalah himpunan titik (x,y,z) di mana fungsi bernilai konstan (f(x,y,z) = 0). Contoh titik pada permukaan: pâ‚ = (2,0,1).

#### Menghitung Gradien di Titik pâ‚ = (2,0,1)

1. Hitung turunan parsial:
   - $\frac{\partial f}{\partial x} = 2x$
   - $\frac{\partial f}{\partial y} = -1$
   - $\frac{\partial f}{\partial z} = -4$

2. Evaluasi di titik (2,0,1):
   - $\frac{\partial f}{\partial x}\bigg|_{(2,0,1)} = 4$
   - $\frac{\partial f}{\partial y}\bigg|_{(2,0,1)} = -1$
   - $\frac{\partial f}{\partial z}\bigg|_{(2,0,1)} = -4$

3. Bentuk vektor gradien:
$$ \nabla f(2,0,1) = 4\mathbf{u}_x - \mathbf{u}_y - 4\mathbf{u}_z \quad \text{(E2.1b.2)} $$

**Interpretasi**:
- Fungsi meningkat paling cepat jika bergerak dalam arah (4,-1,-4)
- Besar laju perubahan: $||\nabla f|| = \sqrt{4^2 + (-1)^2 + (-4)^2} = \sqrt{33}$

#### Menentukan Persamaan Bidang Singgung

Bidang singgung di titik pâ‚ harus:
1. Melewati titik (2,0,1)
2. Tegak lurus terhadap vektor gradien (4,-1,-4)

Persamaan umum bidang singgung:
$$ \nabla f \cdot (\mathbf{r} - \mathbf{r}_0) = 0 $$
$$ \frac{\partial f}{\partial x}(x - x_1) + \frac{\partial f}{\partial y}(y - y_1) + \frac{\partial f}{\partial z}(z - z_1) = 0 $$

Substitusi nilai:
$$ 4(x - 2) + (-1)(y - 0) + (-4)(z - 1) = 0 $$
$$ 4x - 8 - y - 4z + 4 = 0 $$
$$ 4x - y - 4z - 4 = 0 \quad \text{(E2.1b.3)} $$

**Verifikasi**:
- Vektor normal bidang: (4,-1,-4) sama dengan gradien â†’ terbukti tegak lurus

#### Visualisasi (Gambar 2.1c)

- Permukaan 3D $x^2 - y - 4z = 0$
- Bidang singgung $4x - y - 4z = 4$ di titik (2,0,1)
- Vektor gradien (4,-1,-4) yang tegak lurus terhadap bidang singgung

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# --- (a) Tangent line to level curve x^2 - y = 4 ---
p1 = np.array([2, 0])
grad_f = np.array([2*p1[0], -1])  # Gradient at p1

x_vals = np.linspace(p1[0]-1, p1[0]+1, 100)
y_vals = grad_f[1]/grad_f[0]*(x_vals - p1[0]) + p1[1]  # Eq. E2.1a.3

plt.figure(figsize=(15,5))
plt.subplot(131)
plt.plot(x_vals, y_vals, 'r', label='Tangent Line')
x = np.linspace(-3, 3, 400)
y = x**2 - 4
plt.plot(x, y, 'k--', label='x^2 - y = 4')
plt.quiver(p1[0], p1[1], grad_f[0], grad_f[1], angles='xy', scale_units='xy', scale=1)
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.title('(a) Tangent Line to Level Curve')

# --- (b) Using gradient() to plot contour ---
xs = np.linspace(-4, 5, 100)
ys = np.linspace(-4, 5, 100)
Xs, Ys = np.meshgrid(xs, ys)
Fs = Xs**2 - Ys

plt.subplot(132)
contours = plt.contour(Xs, Ys, Fs, levels=[-1, 0, 1, 2, 3, 4, 5])
[dx, dy] = np.gradient(Fs, xs, ys)
plt.quiver(Xs, Ys, dx, dy)
plt.axis('equal')
plt.title('(b) Contour and Gradient Field')
plt.grid(True)

# --- (c) Tangent plane to level surface x^2 - y - 4z = 0 ---
p1 = np.array([2, 0, 1])
grad_f = np.array([2*p1[0], -1, -4])

xs = np.linspace(-3, 0.2, 40)
ys = np.linspace(-4, 0.02, 40)
Xs, Ys = np.meshgrid(xs, ys)
Zs = (Xs**2 - Ys)/4

Xp, Yp = np.meshgrid(np.linspace(1.6, 0.2, 40), np.linspace(-2.0, 0.2, 40))
Zp = (grad_f[0]*(Xp - p1[0]) + grad_f[1]*(Yp - p1[1])) / grad_f[2] + p1[2]  # Eq. E2.1b.3

ax = plt.subplot(133, projection='3d')
ax.plot_surface(Xs, Ys, Zs, alpha=0.5, color='cyan', label='Level surface')
ax.plot_surface(Xp, Yp, Zp, alpha=0.5, color='orange')
ax.quiver(p1[0], p1[1], p1[2], grad_f[0], grad_f[1], grad_f[2], length=0.2, color='black')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('(c) Tangent Plane to Surface')
ax.view_init(15, 15)
plt.tight_layout()
plt.show()
```

# 2.2 FUNGSI VEKTOR

## Definisi Fungsi Vektor

Fungsi vektor (bernilai vektor) 2D dan 3D didefinisikan sebagai:

### 2D:
$$
\mathbf{p}(t) = x(t)\mathbf{u}_{x} + y(t)\mathbf{u}_{y} \tag{2.2.1a}
$$

### 3D:
$$
\mathbf{p}(t) = x(t)\mathbf{u}_{x} + y(t)\mathbf{u}_{y} + z(t)\mathbf{u}_{z} \tag{2.2.1b}
$$

**Karakteristik:**
- Merepresentasikan kurva parametrik di ruang 2D/3D
- Parameter \(t\) biasanya waktu atau parameter lain
- Arah panah menunjukkan arah gerak saat \(t\) bertambah

## Turunan dan Vektor Singgung

Turunan fungsi vektor menghasilkan vektor singgung (tangent vector) pada kurva:

### Untuk 2D:
$$
\mathbf{p}'(t) = \frac{d\mathbf{p}}{dt} = \frac{dx}{dt}\mathbf{u}_{x} + \frac{dy}{dt}\mathbf{u}_{y} \tag{2.2.2a}
$$

### Untuk 3D:
$$
\mathbf{p}'(t) = \frac{d\mathbf{p}}{dt} = \frac{dx}{dt}\mathbf{u}_{x} + \frac{dy}{dt}\mathbf{u}_{y} + \frac{dz}{dt}\mathbf{u}_{z} \tag{2.2.2b}
$$

**Persamaan Garis Singgung** di titik $$\mathbf{p}(t_1):$$
$$
\mathbf{L}(t) = \mathbf{p}(t_1) + \mathbf{p}'(t_1) \cdot (t - t_1)
$$

## Panjang Busur (Arc Length)

Panjang busur kurva dari \(t_0\) ke \(t_1\) dihitung dengan mengintegralkan besar vektor singgung:

$$
l = \int_{t_0}^{t_1} \|\mathbf{p}'(t)\|  dt = \int_{t_0}^{t_1} \sqrt{\left(\frac{dx}{dt}\right)^2 + \left(\frac{dy}{dt}\right)^2 + \left(\frac{dz}{dt}\right)^2}  dt \tag{2.2.3}
$$

### Parameterisasi Panjang Busur
Jika fungsi vektor diubah parameternya menjadi panjang busur \(s\):
$$
s = \int_0^t \|\mathbf{p}'(\tau)\|  d\tau \tag{2.2.4}
$$
- Menghasilkan representasi kurva yang independen terhadap kecepatan
- $$\|\mathbf{p}'(s)\| = 1\ (kecepatan\ konstan\ 1)$$

## Aturan Diferensiasi Fungsi Vektor

Aturan turunan untuk fungsi vektor mirip skalar, dengan penyesuaian:

### Perkalian Skalar-Vektor
$$
\frac{d}{dt}\left[q(t)\mathbf{p}(t)\right] = q'(t)\mathbf{p}(t) + q(t)\mathbf{p}'(t) \tag{2.2.5a}
$$

### Penjumlahan Vektor
$$
\frac{d}{dt}\left[\mathbf{p}_1(t) + \mathbf{p}_2(t)\right] = \mathbf{p}_1'(t) + \mathbf{p}_2'(t) \tag{2.2.5b}
$$

### Perkalian Titik (Dot Product)
$$
\frac{d}{dt}\left[\mathbf{p}_1(t) \cdot \mathbf{p}_2(t)\right] = \mathbf{p}_1'(t) \cdot \mathbf{p}_2(t) + \mathbf{p}_1(t) \cdot \mathbf{p}_2'(t) \tag{2.2.5c}
$$

### Perkalian Silang (Cross Product, 3D)
$$
\frac{d}{dt}\left[\mathbf{p}_1(t) \times \mathbf{p}_2(t)\right] = \mathbf{p}_1'(t) \times \mathbf{p}_2(t) + \mathbf{p}_1(t) \times \mathbf{p}_2'(t) \tag{2.2.5d}
$$


## Contoh: Heliks Melingkar


Fungsi vektor heliks melingkar 3D:
$$
\mathbf{p}(t) = t\mathbf{u}_{x} + \cos(2\pi t)\mathbf{u}_{y} + \sin(2\pi t)\mathbf{u}_{z} \tag{E2.3.1}
$$

### Vektor Singgung
Turunan fungsi:
$$
\mathbf{p}'(t) = \mathbf{u}_{x} - 2\pi\sin(2\pi t)\mathbf{u}_{y} + 2\pi\cos(2\pi t)\mathbf{u}_{z}
$$

Di titik \(t=1\) (\(\mathbf{p}(1)=[1,1,0]^T\)):
$$
\mathbf{p}'(1) = \mathbf{u}_{x} + 0\cdot\mathbf{u}_{y} + 2\pi\mathbf{u}_{z} = [1, 0, 2\pi]^T
$$

### Persamaan Garis Singgung
$$
\mathbf{L}(t) = \begin{bmatrix} 1 \\ 1 \\ 0 \end{bmatrix} + \begin{bmatrix} 1 \\ 0 \\ 2\pi \end{bmatrix} (t - 1)
$$

### Panjang Busur per Putaran (\(t=0\) ke \(t=1\))
$$
s = \int_0^1 \sqrt{(1)^2 + (-2\pi\sin2\pi t)^2 + (2\pi\cos2\pi t)^2}  dt = \int_0^1 \sqrt{1 + 4\pi^2}  dt = \sqrt{1 + 4\pi^2}
$$

## Example

### Contoh 2.3: Garis Singgung dan Panjang Busur Helix Lingkaran

### Fungsi Vektor Helix Lingkaran
Pertimbangkan fungsi vektor 3D berikut:
$$\mathbf{p}(t) = t\mathbf{u}_{x} + \cos(2\pi t)\mathbf{u}_{y} + \sin(2\pi t)\mathbf{u}_{z}$$
*(Persamaan E.2.3.1)*

Fungsi ini merepresentasikan **helix lingkaran** (spiral 3D) dengan karakteristik:
- Komponen $x$: Bergerak linier seiring waktu ($t$)
- Komponen $y$ dan $z$: Membentuk lingkaran di bidang YZ ($\cos$ dan $\sin$)

Visualisasi helix:
- Pada $t=0$: $\mathbf{p}(0) = (0, 1, 0)$
- Pada $t=1$: $\mathbf{p}(1) = (1, 1, 0)$
- Pada $t=0.5$: $\mathbf{p}(0.5) = (0.5, -1, 0)$

---

### (a) Garis Singgung di Titik Tertentu
Kita akan mencari garis singgung di titik $\mathbf{p}(1) = [1, 1, 0]^T$ (saat $t=1$).

#### Langkah 1: Turunan Fungsi Vektor
Turunan pertama $\mathbf{p}(t)$ terhadap waktu:
$$\mathbf{p}'(t) = \frac{d}{dt}\begin{bmatrix} t \\ \cos(2\pi t) \\ \sin(2\pi t) \end{bmatrix} = \begin{bmatrix} 1 \\ -2\pi\sin(2\pi t) \\ 2\pi\cos(2\pi t) \end{bmatrix}$$

#### Langkah 2: Evaluasi di $t=1$
$$\mathbf{p}'(1) = \begin{bmatrix} 1 \\ -2\pi\sin(2\pi) \\ 2\pi\cos(2\pi) \end{bmatrix} = \begin{bmatrix} 1 \\ 0 \\ 2\pi \end{bmatrix}$$

#### Langkah 3: Persamaan Garis Singgung
Persamaan garis singgung di titik $\mathbf{p}(1)$:
$$\mathbf{q}(s) = \mathbf{p}(1) + s \cdot \mathbf{p}'(1) = \begin{bmatrix} 1 \\ 1 \\ 0 \end{bmatrix} + s \begin{bmatrix} 1 \\ 0 \\ 2\pi \end{bmatrix} = \begin{bmatrix} 1 + s \\ 1 \\ 2\pi s \end{bmatrix}$$
*(Persamaan E.2.3.2)*

**Interpretasi Geometris:**
- Vektor singgung $\begin{bmatrix} 1 \\ 0 \\ 2\pi \end{bmatrix}$ menunjukkan arah gerakan helix
- Komponen dominan di arah $x$ dan $z$, tidak ada perubahan di $y$
- Untuk setiap langkah $s$ sepanjang garis singgung:
  - Koordinat $x$ bertambah $s$
  - Koordinat $z$ bertambah $2\pi s$
  - Koordinat $y$ tetap konstan

---

### (b) Panjang Busur Helix
Kita hitung panjang busur dari $t=0$ sampai $t=1$.

#### Konsep Dasar Panjang Busur
Panjang busur dihitung dengan mengintegrasikan kecepatan sepanjang kurva:
$$L = \int_{a}^{b} \|\mathbf{p}'(t)\|  dt$$
dimana $\|\mathbf{p}'(t)\|$ adalah besar kecepatan (panjang vektor turunan).

#### Langkah 1: Hitung Besar Kecepatan
$$\begin{align*}
\|\mathbf{p}'(t)\| &= \sqrt{(1)^2 + [-2\pi\sin(2\pi t)]^2 + [2\pi\cos(2\pi t)]^2} \\
&= \sqrt{1 + 4\pi^2\sin^2(2\pi t) + 4\pi^2\cos^2(2\pi t)} \\
&= \sqrt{1 + 4\pi^2(\sin^2(2\pi t) + \cos^2(2\pi t))} \\
&= \sqrt{1 + 4\pi^2(1)} \\
&= \sqrt{1 + 4\pi^2}
\end{align*}$$

**Catatan Penting:** 
- Hasilnya konstan (tidak bergantung $t$)
- Karena $\sin^2\theta + \cos^2\theta = 1$

#### Langkah 2: Integral Panjang Busur
$$L = \int_{0}^{1} \sqrt{1 + 4\pi^2}  dt = \sqrt{1 + 4\pi^2} \int_{0}^{1} dt = \sqrt{1 + 4\pi^2} \approx 6.3623$$
*(Persamaan E.2.3.3)*

**Interpretasi Fisik:**
- Panjang total lintasan helix dari $t=0$ ke $t=1$ adalah $\sqrt{1+4\pi^2} \approx 6.3623$ satuan
- Nilai eksak: $\sqrt{1+4\pi^2}$ (â‰ˆ 6.3623)
- Setiap satuan waktu ditempuh jarak konstan $\sqrt{1+4\pi^2}$

---

### (c) Parameterisasi Ulang dengan Panjang Busur
Kita nyatakan posisi dalam variabel panjang busur $l$ (jarak dari titik awal).

#### Hubungan $l$ dan $t$
Panjang busur dari $0$ ke $t$:
$$l(t) = \int_{0}^{t} \|\mathbf{p}'(\tau)\|  d\tau = \int_{0}^{t} \sqrt{1 + 4\pi^2}  d\tau = \sqrt{1 + 4\pi^2} \cdot t$$
*(Persamaan E.2.3.4)*

#### Invers Relasi
Dari persamaan di atas, kita dapatkan:
$$t = \frac{l}{\sqrt{1 + 4\pi^2}}$$
*(Persamaan E.2.3.5)*

#### Parameterisasi Ulang Posisi
Substitusi $t$ ke persamaan helix:
$$\mathbf{p}(l) = \frac{l}{\sqrt{1 + 4\pi^2}} \mathbf{u}_x + \cos\left(2\pi \cdot \frac{l}{\sqrt{1 + 4\pi^2}}\right) \mathbf{u}_y + \sin\left(2\pi \cdot \frac{l}{\sqrt{1 + 4\pi^2}}\right) \mathbf{u}_z$$
*(Persamaan E.2.3.6)*

#### Vektor Singgung Satuan
Turunkan terhadap $l$:
$$\mathbf{p}'(l) = \frac{d\mathbf{p}}{dl} = \frac{1}{\sqrt{1 + 4\pi^2}} \begin{bmatrix} 1 \\ -2\pi \sin\left(\frac{2\pi l}{\sqrt{1+4\pi^2}}\right) \\ 2\pi \cos\left(\frac{2\pi l}{\sqrt{1+4\pi^2}}\right) \end{bmatrix}$$
*(Persamaan E.2.3.7)*

# 2.3 KECEPATAN DAN PERCEPATAN

## Posisi, Kecepatan, dan Percepatan

Jika posisi partikel di waktu t dinyatakan oleh vektor posisi:
$$
\mathbf{p}(t) = x(t)\mathbf{u}_{x} + y(t)\mathbf{u}_{y} + z(t)\mathbf{u}_{z} \tag{2.3.1}
$$

Maka:
- **Kecepatan** $$\mathbf{v}(t)$$ adalah turunan pertama posisi terhadap waktu:
$$
\mathbf{v}(t) = \frac{d\mathbf{p}}{dt} = \frac{dx}{dt}\mathbf{u}_{x} + \frac{dy}{dt}\mathbf{u}_{y} + \frac{dz}{dt}\mathbf{u}_{z} \tag{2.3.2}
$$

- **Percepatan** $$\mathbf{a}(t)$$ adalah turunan kedua posisi (atau turunan pertama kecepatan):
$$
\mathbf{a}(t) = \frac{d^2\mathbf{p}}{dt^2} = \frac{d^2x}{dt^2}\mathbf{u}_{x} + \frac{d^2y}{dt^2}\mathbf{u}_{y} + \frac{d^2z}{dt^2}\mathbf{u}_{z} \tag{2.3.3}
$$

## Laju (Speed) dan Hubungannya dengan Panjang Busur

**Laju** didefinisikan sebagai besar kecepatan:
$$
v(t) = \|\mathbf{v}(t)\| = \sqrt{\left(\frac{dx}{dt}\right)^2 + \left(\frac{dy}{dt}\right)^2 + \left(\frac{dz}{dt}\right)^2} \tag{2.3.4}
$$

Laju sama dengan turunan panjang busur terhadap waktu:
$$
v(t) = \frac{dl}{dt} = \|\mathbf{p}'(t)\| \tag{2.3.5}
$$

### Sifat Gerak dengan Laju Konstan
Jika laju partikel konstan $$\|\mathbf{v}\| = c$$, maka:
$$
\mathbf{v}(t) \cdot \mathbf{a}(t) = 0 \tag{2.3.7}
$$
Artinya: **vektor kecepatan dan percepatan saling tegak lurus** saat laju konstan.

## Contoh: Gerak Melingkar Beraturan

Posisi partikel bergerak melingkar di bidang-xy:
$$
\mathbf{p}(t) = R \cos(\omega t) \mathbf{u}_{x} + R \sin(\omega t) \mathbf{u}_{y} \tag{2.3.8}
$$

- **Kecepatan**:
$$
\mathbf{v}(t) = -R\omega \sin(\omega t) \mathbf{u}_{x} + R\omega \cos(\omega t) \mathbf{u}_{y} \tag{2.3.9}
$$

- **Percepatan** (sentripetal):
$$
\mathbf{a}(t) = -R\omega^2 \cos(\omega t) \mathbf{u}_{x} - R\omega^2 \sin(\omega t) \mathbf{u}_{y} = -\omega^2 \mathbf{p}(t) \tag{2.3.10}
$$

Terbukti $$\mathbf{v} \cdot \mathbf{a} = 0$$ (tegak lurus) karena laju konstan.

## Mendapatkan Kecepatan dan Posisi dari Percepatan

Jika percepatan diketahui, kecepatan dan posisi dapat diperoleh dengan integrasi:

- **Kecepatan** dari percepatan $$dengan\ kondisi\ awal\ \mathbf{v}(t_0):$$
$$
\mathbf{v}(t) = \mathbf{v}(t_0) + \int_{t_0}^{t} \mathbf{a}(\tau)  d\tau \tag{2.3.12}
$$

- **Posisi** dari kecepatan (dengan kondisi awal \(\mathbf{p}(t_0)\)):
$$
\mathbf{p}(t) = \mathbf{p}(t_0) + \int_{t_0}^{t} \mathbf{v}(\tau)  d\tau \tag{2.3.13}
$$

### Interpretasi Fisik:
1. **Percepatan** â†’ Perubahan kecepatan per satuan waktu
2. **Kecepatan** â†’ Perubahan posisi per satuan waktu
3. **Laju Konstan** â†’ Percepatan hanya mengubah arah kecepatan (tidak mengubah besarnya)
4. **Gerak Melingkar** â†’ Percepatan sentripetal selalu menuju pusat lingkaran

## Example

### Bagian 1: Gerak Peluru (Contoh 2.4)

#### **Konsep Dasar**
Gerak peluru merupakan contoh gerak parabola akibat percepatan gravitasi. Gerak ini dapat dianalisis dengan memisahkan komponen horizontal (x) dan vertikal (z).

**Data Awal:**
- Kecepatan awal, \( v_0 = 250  \text{m/s} \)
- Sudut elevasi, \( \theta_0 = 30^\circ \)
- Ketinggian awal, \( z_0 = 2  \text{m} \)
- Percepatan gravitasi, \( g = 9.8  \text{m/s}^2 \)

---

#### **Langkah-langkah Solusi**

##### 1. **Kondisi Awal Vektor**
- Posisi awal: 
  $$
  \vec{p}(0) = 0 \hat{u}_x + 2 \hat{u}_z
  $$

- Kecepatan awal:
  $$
  \vec{v}(0) = (250 \cos 30^\circ) \hat{u}_x + (250 \sin 30^\circ) \hat{u}_z
  $$
  $$
  \vec{v}(0) = 216.5 \hat{u}_x + 125 \hat{u}_z  \text{m/s}
  $$

- Percepatan (konstan):
  $$
  \vec{a}(t) = -9.8 \hat{u}_z  \text{m/s}^2
  $$

---

##### 2. **Persamaan Kecepatan**
Diperoleh dengan mengintegralkan percepatan:
$$
\vec{v}(t) = \vec{v}(0) + \int_0^t \vec{a}(t) dt = 216.5 \hat{u}_x + (125 - 9.8t) \hat{u}_z
$$

**Interpretasi:**
- Komponen x konstan (tidak ada percepatan horizontal)
- Komponen z berkurang linear karena gravitasi

---

##### 3. **Persamaan Posisi (Lintasan)**
Diperoleh dengan mengintegralkan kecepatan:
$$
\vec{p}(t) = \vec{p}(0) + \int_0^t \vec{v}(t) dt
$$
$$
x(t) = 216.5 t
$$
$$
z(t) = 2 + 125t - 4.9t^2
$$

**Bentuk lintasan:** Parabola (kombinasi gerak GLB horizontal dan GLBB vertikal)

---

##### 4. **Waktu Tempuh Hingga Mendarat (\( t_f \))**
Peluru mendarat saat \( z(t_f) = 0 \):
$$
2 + 125t_f - 4.9t_f^2 = 0
$$

**Solusi persamaan kuadrat:**
$$
t_f = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a} \quad \text{dengan} \quad a = -4.9,  b = 125,  c = 2
$$
$$
t_f = \frac{-125 \pm \sqrt{125^2 - 4(-4.9)(2)}}{2(-4.9)} = \frac{-125 \pm \sqrt{15625 + 39.2}}{-9.8}
$$
$$
t_f = \frac{-125 \pm 125.16}{-9.8}
$$

**Akar fisik (positif):** 
$$
t_f = \frac{-125 - 125.16}{-9.8} \approx 25.53  \text{detik}
$$

---

##### 5. **Jangkauan (Range)**
Substitusi \( t_f \) ke persamaan x:
$$
x_f = 216.5 \times 25.53 \approx 5527  \text{meter}
$$

---

##### 6. **Kecepatan Saat Tumbukan**
- Komponen horizontal: \( v_x = 216.5  \text{m/s} \) (konstan)
- Komponen vertikal: 
  $$
  v_z = 125 - 9.8 \times 25.53 \approx -125.2  \text{m/s}
  $$
  
**Besar kecepatan:**
$$
v_f = \sqrt{(216.5)^2 + (-125.2)^2} = \sqrt{46872.25 + 15675.04} = \sqrt{62547.29} \approx 250.1  \text{m/s}
$$

**Fakta menarik:** \( v_f \approx v_0 \) karena energi mekanik hampir kekal (perbedaan ketinggian kecil).

### Bagian 2: Konsep Kelengkungan (Curvature)

#### **Definisi dan Signifikansi**
Kelengkungan (\(\kappa\)) mengukur **seberapa tajam sebuah kurva berbelok** pada suatu titik. Semakin tajam belokan, semakin besar \(\kappa\).

---

#### **Konsep Kunci**

##### 1. **Vektor Tangen Satuan (\(\mathbf{u_T}\))**
Vektor yang menyinggung kurva dan memiliki panjang 1:
$$
\mathbf{u_T}(t) = \frac{\vec{v}(t)}{\|\vec{v}(t)\|} = \frac{\vec{v}(t)}{v(t)}
$$

**Contoh:** Pada gerak peluru, \(\mathbf{u_T}\) selalu searah dengan arah gerak sesaat.

---

##### 2. **Panjang Busur (\(l\))**
Jarak yang ditempuh sepanjang kurva:
$$
\frac{dl}{dt} = v(t)
$$

---

##### 3. **Definisi Matematis Kelengkungan**
$$
\kappa = \left\| \frac{d\mathbf{u_T}}{dl} \right\| = \frac{\| d\mathbf{u_T}/dt \|}{dl/dt} = \frac{\| \mathbf{u_T}'(t) \|}{v(t)}
$$

**Interpretasi:** Laju perubahan arah gerak per satuan panjang busur.

---

##### 4. **Radius Kelengkungan (\(r_\kappa\))**
Jari-jari lingkaran yang paling sesuai dengan bentuk kurva di suatu titik:
$$
r_\kappa = \frac{1}{\kappa}
$$

**Lingkaran kelengkungan:** Lingkaran dengan radius \(r_\kappa\) yang menyentuh kurva di titik tersebut.

---

#### **Aplikasi pada Gerak Peluru**

##### Rumus Khusus Gerak Peluru:
$$
\kappa(t) = \frac{g \cdot |v_x|}{[v_x^2 + v_z(t)^2]^{3/2}} = \frac{g \cdot |v_x|}{v(t)^3}
$$

##### Analisis Titik Penting:
1. **Titik Awal (\(t=0\)):**
   - $$ v(0) = 250  \text{m/s} $$
   - $$ \kappa = \frac{9.8 \times 216.5}{250^3} \approx 0.000136  \text{m}^{-1}$$
   - $$ r_\kappa \approx 7353  \text{m}\ (kelengkungan\ kecil)$$

2. **Puncak Lintasan (\(v_z=0\)):**
   - $$ v = v_x = 216.5  \text{m/s} $$
   - $$ \kappa = \frac{9.8 \times 216.5}{(216.5)^3} \approx 0.000209  \text{m}^{-1} $$
   - $$ r_\kappa \approx 4784  \text{m}\ (kelengkungan\ maksimum)$$

3. **Titik Tumbukan (\(t=t_f\)):**
   - $$Mirip\ titik\ awal\ ( v \approx 250  \text{m/s} )$$
   - $$ \kappa \approx 0.000136  \text{m}^{-1} $$
