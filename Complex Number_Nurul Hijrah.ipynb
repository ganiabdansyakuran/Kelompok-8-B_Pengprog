## 3.1 Compleks Number (Bilangan Kompleks)

Bilangan kompleks adalah gabungan antara **bilangan riil** dan **bilangan imajiner**. Bentuk umum bilangan kompleks dinyatakan sebagai:

$$
z = a + bi \tag{1}
$$

di mana:

- **a** adalah bagian *riil*  
- **b** adalah bagian *imajiner*  
- **i** adalah bilangan imajiner yang memenuhi sifat:
$$
ùëñ = \sqrt{-1} \tag{2}
$$

Dalam sistem bilangan real biasa, kita tidak bisa menghitung akar dari bilangan negatif, sehingga bilangan kompleks diciptakan untuk memperluas sistem bilangan tersebut. Dengan adanya bilangan kompleks, kita dapat menyelesaikan persamaan matematika yang sebelumnya tidak memiliki solusi dalam bilangan real, seperti  $ x^2 + 1 = 0 $.

Salah satu alasan mengapa bilangan ini disebut sebagai **bilangan kompleks** adalah karena adanya bagian imajiner yang dilambangkan dengan huruf **$ i $**. 

**Contoh:**
$$
z = 4+3i \tag{3}
$$
Nah pada Bilangan di atas tersebut adalah bilangan yang komplekss yang mana bagian riil nya adalah $ 4 $ dan bagian imajiner nya ada  $ 3i $.

Selain berguna dalam matematika, bilangan kompleks juga banyak digunakan dalam bidang lain seperti fisika, teknik elektro, dan pemrosesan sinyal, karena kemampuannya dalam merepresentasikan fenomena yang melibatkan gelombang, rotasi, dan osilasi

## 3.2 Operasi Bilangan Kompleks

Pada sistem bilangan komplek ada beberapa operasi di dalam nya ialah operasi perkalian, penjumlahan, pembagian, pengurangan, dan lain sebagainya.

Berikut penjelasan dari operasi penjumlahan dan pengurangan bagiannya.

- ### Operasi Pengurangan

Operasi pengurangan pada bilangan ini dapat di lakukan dengan cara mengelompokan bagian bilangan riil dan bilangan imajiner yang kemudian akan di lakukan pengurangan pada bagian masing-masing.
$$ 
(a + bi) - (c + di) = (a - c) + (b -d)i \tag{4}
$$ 

**Contoh operasi pengurangan:**
$$
(6 + 3i) - (3 - 2i) 
$$
**Jawab:**

   $$6 - 3 = 3$$  
   $$3i - (-2i) = 3i + 2i = 5i$$
$$(6 + 3i) - (3 - 2i) = 3 + 5i$$


-  ### Operasi Penjumlahan

Operasi penjumlahan dalam bilangan ini, yang mana penjumlahannya dapat di lakukan dengan melakukan pengelompok kan bagian bilangan riil dan bagian bilangan imajiner yang lalu akan di jumlahkan masing-masing.
$$ 
(a + bi) + (c + di) = (a + c) + (b + d)i \tag{5}
$$

**Contoh:**
$$ (7+3i) + (4-1i) $$

**Jawab:**

$$7 + 4 = 11$$
$$3i - 1i = 2i$$
$$(7 + 3i) + (4 - 1i) = 11 + 2i$$

- ### Operasi Perkalian
Perkalian itu seperti menyebarkan dua pasangan bilangan (real dan imajiner) dan menggabungkannya jadi satu.

$$
(a + bi)(c + di) = (ac - bd) + (ad + bc)i \tag{6}
$$

**Contoh:**

$$
(2 + 3i)(4 + 5i) = (2 \cdot 4 - 3 \cdot 5) + (2 \cdot 5 + 3 \cdot 4) i
$$

**Jawab:**

$$ 
(2 + 3i)(4 + 5i) = (2 \cdot 4 - 3 \cdot 5) + (2 \cdot 5 + 3 \cdot 4) i
$$

$$ 
= (8 - 15) + (10 + 22)i 
$$

$$ 
= -7 + 22i
$$


- ### Operasi Pembagian
Pembagian itu seperti merapikan hasil agar tidak ada bilangan imajiner di penyebut, supaya bentuk akhirnya lebih "standar"

$$
\frac{a + bi}{c + di} = \frac{(ac + bd)+(bc-ad)i}{c^2 + d^2} \tag{6}
$$

**Contoh:**
$$
\frac{2 + 3i}{1 + 2i} = \frac{(2 \cdot 1 + 3 \cdot 2) + (3 \cdot 1 - 2 \cdot 2)i}{1^2 + 2^2}
$$

$$
= \frac{(2 \cdot 1 + 3 \cdot 2) + (3 \cdot 1 - 2 \cdot 2)i} {1^2 + 2^2}
$$
$$
= \frac{(2 + 6) (3 - 4)i}{1 + 4}
$$
$$
= \frac{8 - i}{5}
$$
$$
= \frac{8}{5} - \frac{1}{5}i
$$

### 3.3 Konjugat Kompleks

Konjugat kompleks dari suatu bilangan kompleks adalah bilangan yang diperoleh dengan mengganti tanda bagian imajinernya. Bilangan kompleks berguna untuk menyederhanakan pembagian dan perhitungan modulus.

Apabila anda punya bilangan kompleks:
$$
z = a + bi
$$

Maka, **konjugat kompleksnya** adalah:
$$
\bar{z} = a -bi
$$ 

Contoh: jika  $ \bar{z} = 4 + 5i $, maka konjugat kompleksnya adalah:

$$
\bar{z} = 4 - 5i
$$

#### 3.3.1 Penyederhanaan Pembagian
Apabila  anda ingin membagi bilangan komppleks, anda dapat mengalikan pembilang dan penyebut dengan konjugat dari penyebut.
Contoh:
$$
\frac{2 + 3i}{4 - i}
$$

Kalikan dengan konjugat dari penyebut $(4 + i)$:
$$
\frac{2 + 3i}{4 - i} \cdot \frac{4 + i}{4 + i} = \frac{(2 + 3i)(4 + i)}{(4-i)(4+i)}
$$
Ini akan membuat penyebut menjadi bilangan real.

#### 3.3.2 Perhitungan Modulus (jarak dari titik ke pusat)
Jika $z = a + bi$, maka modulusnya:
$$
|z| = \sqrt{a^2 + b^2}
$$

Dan ini juga bisa dihitung dengan menggunakan:
$$
z \cdot \bar{z} = (a + bi)(a - bi) = a^2 + b^2
$$

karena 
$|z| = \sqrt{z \cdot \bar{z}}$

## 3.4 Pangkat dan Pola Pangkat $i$
Dalam bilangan kompleks, $i$ adalah unit imajiner, yang didefinisikan sebagai:
$$
i = \sqrt{-1}
$$

Jika  menaikkan $i$ ke pangkat  yang  lebih tinggi, akan menemukan pola berulang setiap 4 langkah.

- **Pola Pangkat $i$**:

$i^1 = i$

$i^2 = -1$

$i^3 = -1$

$i^4 = 1$

$i^5 = i$ (karena kembali ke pola awal)

$i^6 = -1$

$i^7 = -1$

$i^8 = 1$ (dan seterusnya...)

Contoh:
Tentukan nilai  dari:
$$
i^{11} + i^5
$$

Jawab:
$i^{11} \Rightarrow 11$ mod = $4$ $\Rightarrow$ $i^{11} = -1$

$i^5 \Rightarrow 5$ mod = $4$ $\Rightarrow$ $^5 = i$

$\Rightarrow$ $-i + i = 0$

## 3.5 Representasi Geometris
Bilangan kompleks $Z = a + bi$ bisa direpresentasikan sebagai titik di bidang datar, yang disebut **Bidang Argand**.

Anggap bilangan kompleks seperti titik koordinat di bidang 2 dimensi:
$$
z = a + bi \rightarrow (a,b)
$$

- a = bilangan real, diletakkan di sumbu horizontal (x-axis)
- b = bilangan imajiner,  diletakkan di sumbu vertikal (y-axis)

Misalnya:
$$
z = 3 + 2i
$$
Maka di bidang Argand, tandai titik (3,2)
- 3 langkah ke kanan (karena real = 3)
- 2 langkah ke atas (karena imajiner = 2)

Contohnya:
Gambarkan posisi dan arah vektor dari bilangan kompleks berikut di bidang Argand:
1. $z1 = 3 + 2i$
2. $z2 = -2 + 4i$
3. $z3 = -1 - 3i$
4. $z4 = 2 - i$

## Bentuk Polar bilangan Kompleks

Kita juga dapat mempresentasikan bilangan kompleks bentuk  polar dalam pasangan terurut ($r$, $\theta$). Bilangan kompleks dapat dinyatakan dalm bentuk polar sebagai:
$$
z = a + bi = r \cos (\theta) + ir \sin (\theta)
$$
Kadanga-kadang, juga dapat dituliskan dengan bentuk eksponensial:
$$
z = re^{i \theta} = r [cos (\theta) + i \sin (\theta)]
$$
